## 2.2 match和search
不是返回字符串列表，而是一个`MatchObject`，可以获得更多信息。

如果匹配失败，会返回一个`NoneType`。
- match 从字符串开头开始匹配，如果开头位置没有匹配成功就算失败
- search 跳过开头，继续向后寻找是否有匹配的字符串

````python
input2 = '123NLP'
````

````python
pattern = re.compile(r'\d')
match = re.match(pattern, input2)
match.group()
````

## 2.3 字符串的替换和修改
sub(rule, replace, target, \[count\])

subn(rule, replace, target, \[count\])

rule : 正则规则

replace : 替换用的字符串

target : 目标字符串，被替换的字符串

count : 最多替换次数

----
subn 会返回一个元组，第一个元素是被替换的字符串（同sub函数），第二个元素是替换的次数

````python
pattern = re.compile(r'\d')
re.sub(pattern, '|_|', input2, 3)
# re.sub(pattern, '|_|', input2)
````

````python
pattern = re.compile(r'\d')
re.subn(pattern, '|_|', input2) # subn的返回
````

````python
pattern = re.compile(r'\d{3}')
re.subn(pattern, '【3个数字】', input2) # subn的返回
````

### 分块

依据定义的规则（例如下面代码的grammer，正则表达）在句子中去抽取一些部分

````python
from nltk.chunk import RegexpParser
````

````python
sentence = [ ('the','DT'), ('little','JJ'), ('yellow','JJ'), ('dog','NN'), ('died','VBD') ]
grammer = "MY_NP: {<DT>?<JJ>*<NN>}"
# 正则 MY_NP:我的块   规则：<限定词>?<形容词>*<名词>

cp = nltk.RegexpParser(grammer) # 生成规则
result = cp.parse( sentence )

print(result)

result.draw()
````

# 4. 数据清洗

## 数据清洗实例

````python
import re
from nltk.corpus import stopwords
````

````python
# input data
s = '  RT @Amila #Test \nTom\'s newly listed Co &amp; Mary\'s unlisted   \
        Group to supply tech for nlTK.\nh $TSLA $AASL http://t.co/x34afsfQsh'
s
````

````python
# 指定停用词
cache_english_stopwords = stopwords.words('english')
````

文本清洗

替换方法：

sub(rule, replace, target, [count])

subn(rule, replace, target, [count])

````python
def text_clean(text): # 文本清洗
    print('原始数据：', text, '\n')
    
    #去掉HTML标签
    #例如： &amp;
    html_pattern = re.compile(r'\&\w*;|#\w*|@\w*')
    text_no_sepcial_entities = re.sub(html_pattern, '',text)
    #text_no_sepcial_entities = re.sub(r'\&\w*;|#\w*|@\w*', '', text)
    print('去掉html标签：', text_no_sepcial_entities, '\n')
    
    #去掉$符号
    tickers_pattern = re.compile(r'\$[a-zA-Z]*')
    text_no_tickers = re.sub(tickers_pattern, '', text_no_sepcial_entities)
    print('去掉$标签：', text_no_tickers, '\n')
    
    #去掉超链接
    hyper_link_pattern = re.compile(r'http:\/\/.*\/\w*')
    text_no_hyper_link = re.sub(hyper_link_pattern, '', text_no_tickers)
    print('去掉超链接：', text_no_hyper_link, '\n') 
    
    #去掉专有名词符号
    #即一些只有一两个字母的缩写单词
    small_word_pattern = re.compile(r'\b\w{1,2}\b') # \b代表单词边界
    text_no_small_word = re.sub(small_word_pattern, '', text_no_hyper_link)
    print('去掉小单词：', text_no_small_word, '\n') 

    #去除多余空格
    space_pattern = re.compile(r'\s+')
    text_no_more_space = re.sub(space_pattern, ' ', text_no_small_word)
    text_no_more_space = text_no_more_space.strip(' ')  # 这句话好像失效了
    print('去掉多余空格：', text_no_more_space, '\n')
    
    #分词
    tokens = word_tokenize(text_no_more_space)
    print("分词结果：", tokens, '\n')
    
    #去掉停用词
    list_no_stopwords = [word for word in tokens if word not in cache_english_stopwords]
    print("去掉停用词：", list_no_stopwords, '\n')
    
    #整合为一个字符串
    text_filtered = ' '.join(list_no_stopwords) #需要用空格
    text_filtered = text_filtered.lstrip(' ') #为什么还有一个空格删不掉
    print("过滤后：", text_filtered, '\n')
````

````python
text_clean(s)
````
